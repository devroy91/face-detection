```python
##importing required libs
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
##converting from BGR to RGB, 
##since opencv opens an image in BGR format
def convertToRGB(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
def detectFaceHaar(img):
    ##load image
    img1 = cv2.imread(img)

    ##convert it to greyscale
    gray_img = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

    plt.imshow(gray_img, cmap = 'gray')

    ##load haar cascade
    haar_classifier = cv2.CascadeClassifier('lib/haarcascade_frontalface_alt.xml')

    #detecting faces (some images may be closer to camera than others) images 
    faces = haar_classifier.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=5)

    #print the number of faces found 
    print('Faces found: ', len(faces))

    ##draw a rectangle around faces
    for (x, y, w, h) in faces:

        print(x, y, w, h)
        cv2.rectangle(img1, (x, y), (x+w, y+h), (0, 255, 0), 2)


    ##convert img to rgb and show image
    plt.imshow(convertToRGB(img1))
    plt.show()
```

```python
detectFaceHaar('data/img1.jpg')
```
